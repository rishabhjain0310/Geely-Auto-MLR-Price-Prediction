```python
# Creating and Dependent and independent Variable.
# As we are using Linear Regression Algorithm we cannot include column that has object in it.

x = df[['wheelbase','carlength','carwidth','curbweight','enginesize','boreratio',
      'horsepower','citympg','highwaympg']]
```

```python
x
```

```python
y = df.price
```

```python
y
```

```python
# Model Training and Testing.

from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=42)
```

```python
# Call Algorithm 

from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr
```

```python
lr.fit(x_train,y_train)
```

```python
y_pred = lr.predict(x_test)
```

```python
y_pred
```

```python
# Model Evaluation 

from sklearn.metrics import r2_score
```

```python
r2_score(y_test,y_pred)
```

```python
data = pd.DataFrame({'Original':y_test,'Predict':y_pred,'Difference':y_test-y_pred})
```

```python
data
```

```python
sns.regplot(x=data.Original,y=data.Predict)
```
